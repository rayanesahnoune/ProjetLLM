```python
import numpy as np 

def softmax_row(row):
    row_exp=np.exp(row)
    return row_exp/np.sum(row_exp)
#n=10 
#d=8
#X=np.random.randn(n,d)
#B=softmax_row(X[0])
#print (B)
```

```python
n=10 
d=8
X=np.random.randn(n,d)
S=X@X.T
A=np.zeros((n,n))
for i in range(n):
    A[i]=softmax_row(S[i])
O=A@X
print("\nTaille (n*d) des matrices X et O respectivement ")
print(X.shape,O.shape)
print("\nNombres d'élements dans S et A  ")
print(S.size,A.size)
print("\nAffichage d'une ligne de la matrrice A ,on voit que toutes les valeurs sont cmprises entre 0 et 1   ")
print(A[3])
print("\nTaille de la matrice O et taille d'une ligne de X")
print(O.size,X[0].size)
```
