```python
import numpy as np
import tensorflow as tf
class SimpleAttention(tf.keras.layers.Layer):
    def __init__(self):
         # obligatoire
         super().__init__()
         # ici : définir les attributs self.S, self.A, self.O (pour pouvoir les afficher...
         self.A=None
         self.O=None
         self.S=None
         
    def call(self, X):
     # recopier ici les calculs de l’étape précédente :
          S=X @ tf.transpose(X)
     # 2) A = ... (softmax ligne par ligne)
          n=X.shape[0]
          A=np.zeros((n,n))
          for i in range(n):
               valeurs_exp = np.exp(S[i])
               sommeExp = np.sum(valeurs_exp) 
               A[i]= valeurs_exp / sommeExp
     
          O=A@X
          self.A=A
          self.O=O
          self.S=S
          return O
    
    def get_S(self):
     return self.S
    
    def get_A(self):
     return self.A
    
    def get_O(self):
     return self.O
X = np.random.randn(10, 8)
layer = SimpleAttention()
O = layer(X)

print(f"Taille de X : {X.shape}")
print(f"Taille de O : {O.shape}")
print(f"Ligne 3 de A : {layer.get_A()[3]}")
print(f"Somme ligne 3 : {np.sum(layer.get_A()[3])}")
```
