```python
import numpy as np
 
class SimpleAttention():
    def __init__ (self):
       pass
    def forward( self,X):
        S=(X @ X.T) 
        n=X.shape[0]
        A=np.zeros((n,n))
        for i in range(n):
            valeurs_exp = np.exp(S[i])
            sommeExp = np.sum(valeurs_exp) 
            A[i]= valeurs_exp / sommeExp
        O=A@X
        return O,A
X = np.random.randn(10,8)
att = SimpleAttention()
O, A = att.forward(X)
print("=== VÉRIFICATION DES TAILLES ===")
print(f"Taille de X  : {X.shape}")
print(f"Taille de O : {O.shape}") 
print("\n=== CONTENU DE La ligne 3 DE LA MATRICE A")
print("Ligne 3",A[3])
print(f"Somme de la ligne 3 : {np.sum(A[3])}")
print("===COMPARAISON ENTRE DEUX LIGNES DE X ET O (Ligne 9)  ===")
print(f"Ligne X    :\n{X[ 9]}")
print(f"Ligne O   :\n{O[9]}")
```
